```python
#remove duplicates
nums = [1, 2, 2, 3, 1, 4, 5, 3]
res = set(nums)
print(res)
```

```python
#Replace None values with the average of the valid numbers.
nums = [10, None, 20, None, 30]

valid_nums = []

for x in nums:
    if x is not None:
        valid_nums.append(x)
print(valid_nums)

average = sum(valid_nums) / len(valid_nums)

result = []

for x in nums:
    if x is None:
        result.append(int(average))
    else:
        result.append(x)
print(result)
```

```python
#remove negative values from list
values = [10, -3, 25, -1, 7]
res=[]
for x in values:
    if x > 0:
        res.append(x)
print(res)
```

```python
#Multiply each value by 2
features = [1, 2, 3, 4]

res =[]

for x in features:
    res.append(x*2)

print(res)
```

```python
#Create pairs of consecutive values.
values = [5, 10, 15, 20]

result =[]
for x in range(len(values)-1):
    result.append((values[x],values[x+1]))
print(result)
```
