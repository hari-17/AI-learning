```python
#Replace None values with the average of the valid numbers.
nums = [10, None, 20, None, 30]

valid_nums = []

for x in nums:
    if x is not None:
        valid_nums.append(x)
print(valid_nums)

average = sum(valid_nums) / len(valid_nums)

result = []

for x in nums:
    if x is None:
        result.append(int(average))
    else:
        result.append(x)
print(result)
```

```python
#Count how many times each label appears.
labels = ["A", "B", "A", "C", "A"]
hm = {}
for x in labels:
    if x in hm:
        hm[x] = hm[x] + 1
    else:
        hm[x] = 1

print(hm)
```

```python
#Create a dictionary by combining two lists.
features = ["age", "salary"]
values = [30, 50000]

hm = {}

for i in range(len(features)):
    hm[features[i]] = values[i]

print(hm)
```

```python
#Group values by label.
data = [("A", 90), ("B", 85), ("A", 88)]

hm = {}

for alpha, number in data:
    if alpha in hm:
        hm[alpha].append(number)
    else:
        hm[alpha] = [number]

#or

res = {}

for alpha, number in data:
    res.setdefault(alpha,[]).append(number)

print(res)


print(hm)
```

```python
#Remove duplicate values and count frequency.
values = [1, 2, 2, 3, 1, 4]

hm ={}
for x in values:
    if x in hm:
        hm[x] = hm[x] +1
    else:
        hm[x] = 1

print(hm)
```

```python
#Count only positive numbers.
values = [1, -1, 2, -2, 3, 1]

hm={}

for x in values:
    if x > 0:
        if x in hm:
            hm[x] += 1
        else:
            hm[x] = 1
print(hm)
```

```python
#Convert raw records into a structured dictionary.
records = [
    ("A", 90),
    ("B", 85),
    ("C", 92)
]
res = {}
for x,y in records:
    res[x] = {"score":y}
print(res)
```

```python
#From a list of words, build a vocabulary and map each word to its frequency.
words = ["ai", "ml", "ai", "dl", "ml", "ai"]

hm ={}

for key in words:
    if key in hm:
        hm[key] += 1
    else:
        hm[key] = 1
print(hm)
```

```python
#Convert raw data into model-ready format.
data = [
    ("age", 30),
    ("salary", 50000),
    ("experience", 5)
]
hm = {}
hm["Feature"] = []
hm["Feature_Name"] = []
for key,value in data:
    hm["Feature"].append(key)
    hm["Feature_Name"].append(value)

print(hm)
```
